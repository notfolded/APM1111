4. Verifying the relationship between moments of the Normal Data set

> a. $m_2 = m_2' - m_1'^2 $

> b. $m_3 = m_3' - 3m_1'm_2' + 2m_1'^3 $

> c. $m_4 = m_4' - 4m_1'm_3' + 6m_1'^2m_2'- 3m_1'^4$

```python
m2 = secondmoment - (firstmoment*firstmoment)
m3 = thirdmoment - 3*firstmoment*secondmoment + 2*(firstmoment**3)
m4 = fourthmoment - 4*firstmoment*thirdmoment + 6*(firstmoment*firstmoment*secondmoment) - 3*(firstmoment**4)

print("Moments About the mean of Normal")
print(f"first moment: {normal_meanmoment}")
print(f"second moment: {normal_secondmean}")
print(f"third moment: {normal_thirdmean}")
print(f"fourth moment: {normal_fourthmean}")

print("\nUsing the Given Formula for verification we get:")
print(f"\na) m2: ({secondmoment})-({firstmoment})^2 = {m2}")
print(f"b) m3: ({thirdmoment}) - 3({firstmoment})({secondmoment}) + 2({firstmoment})^3 = {m3}")
print(f"c) m4:  ({fourthmoment}) - 4({firstmoment})({thirdmoment}) + 6({firstmoment})^2({secondmoment}) - 3({firstmoment})^4 = {m4}")
```
